```python
import types
import json
import os
import spotipy
import spotipy.util as util
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
## User needs to generate a client secret and ID's keys for access to this application
CLIENT_SECRET = os.environ.get('SPOTIPY_CLIENT_SECRET')
CLIENT_ID = os.environ.get('SPOTIPY_CLIENT_ID')
USERNAME = 'raydeezy'
REDIRECT_URI = 'https://127.0.0.1/'
```

```python
scopes = "user-read-recently-played user-read-private user-top-read"

token = util.prompt_for_user_token(
        username=USERNAME,
        scope=scopes,
        client_id=CLIENT_ID,
        client_secret=CLIENT_SECRET,
        redirect_uri=REDIRECT_URI  
)

sp = spotipy.Spotify(auth=token)
```

```python
rp = sp.current_user_recently_played()
```

```python
track_ids = []

rp_json = json.loads(json.dumps(rp))
for x in rp_json['items']:
    track_ids.append(x['track']['uri'])
```

```python
def getAudioFeatures(track_ids): 
    
    audioFeatures = []
    
    for id in track_ids:
        features = sp.audio_features(id)[0]
        audioFeatures.append(features)
    
    return audioFeatures
```

```python
audioFeatures = getAudioFeatures(track_ids)
```
